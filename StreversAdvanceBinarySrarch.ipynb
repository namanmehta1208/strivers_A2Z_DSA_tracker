```python
# Search in rotated sorted array-II
class Solution:
    def __init__(self, nums = None, k = 0):
        self.nums = nums if nums is not None else []
        self.k = k

    def searchInARotatedSortedArrayII(self):
        i, j = 0, len(self.nums) - 1

        print(f"i = {i}, j = {j}")

        while(i < j):

            mid = (i + j) // 2
            print(f"i = {i}, j = {j}")
            print(f"mid = {mid}")

            if(self.nums[mid] == self.k):
                return True

            if(self.nums[i] == self.nums[mid] == self.nums[j]):
                i += 1
                j -= 1
                continue

            if(self.nums[mid] <= self.nums[j]):
                if(self.nums[mid] <= self.k <= self.nums[j]):
                    i = mid + 1
                else:
                    j = mid - 1
            else:
                if(self.nums[i] <= self.k <= self.nums[mid]):
                    j = mid - 1
                else:
                    i = mid + 1


        return False
```

```python
sol = Solution([38,53,53,53,54,-92,-92,-92,-75,-70,-59,-59,-59,-51,-44,-13,-13,22,22,22], -13)

sol.searchInARotatedSortedArrayII()
```

```python
# Find out how many times the array is rotated
"""
- first we check if the array is already sorted, if yes then we return the start, as it will be smallest
- Next we check which half is sorted, if right half is sorted mid element will be the smallest and we keep it if it is smaller then our ans, and eleminate the right half as we got the smallest from there
- If left half is sorted, start element will be the smallest, after checking with ans we will keep it if it is smaller.
"""
def findKRotation(nums):
    start, end = 0, len(nums) - 1
    index = float('inf')
    ans = float('inf')

    while(start <= end):
        mid = (end + start) // 2
        print(f"start = {start}, mid = {mid}, end = {end}")
        print(f"ans = {ans}, index = {index}")

        if(nums[start] < nums[end]):
            if(nums[start] < ans):
                ans = nums[start]
                index = start
            break

        if(nums[mid] < nums[end]):
            if(nums[mid] < ans):
                ans = nums[mid]
                index = mid
            end = mid - 1
        else:
            if(nums[start] < ans):
                ans = nums[start]
                index = start
            start = mid + 1
        print(f"start = {start}, mid = {mid}, end = {end}")
        print(f"ans = {ans}, index = {index}")
        print("--------------------------------------")

    return index
```

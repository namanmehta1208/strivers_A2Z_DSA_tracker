```python
# Min Heap algorithm
"""
Here we have two scnario, heapify down, if parent is greater then it's child
heapify up if child is minimum then the parent

so first we are updating the array and we are checking for which scanario to follow
heapify up or heapify down and so on!!!
"""
class Solution:
    def heapify(self, nums, ind, val):
        n = len(nums)

        if(n == 1):
            nums[0] = val
            return nums

        def heapifyDown(index):
            # print(f"heapifying down called")
            print(f"Initial index : {index}, n : {n}")

            while(index < n):
                parentIndex = index
                lChild = (2 * index) + 1
                rChild = (2 * index) + 2

                print(f"left child : {lChild}, rChild : {rChild}")

                if(lChild >= n and rChild >= n):
                    break
                elif(lChild < n and nums[lChild] > nums[index] and rChild < n and nums[rChild] > nums[index]):
                    break

                # if(lChild < n and nums[lChild] < nums[parentIndex] and rChild < n and nums[rChild] > nums[lChild]):
                #     nums[lChild], nums[parentIndex] = nums[parentIndex], nums[lChild]
                #     index = lChild
                # elif(rChild < n and nums[rChild] < nums[parentIndex] and lChild < n and nums[lChild] > nums[rChild]):
                #     nums[rChild], nums[parentIndex] = nums[parentIndex], nums[rChild]
                #     index = rChild
                # else:
                #     break

                if(lChild < n and nums[lChild] < nums[parentIndex]):
                    if(rChild < n and nums[rChild] < nums[lChild]):
                        nums[rChild], nums[parentIndex] = nums[parentIndex], nums[rChild]
                        index = rChild
                    else:
                        nums[lChild], nums[parentIndex] = nums[parentIndex], nums[lChild]
                        index = lChild

                elif(rChild < n and nums[rChild] < nums[parentIndex]):
                    if(lChild < n and nums[lChild] < nums[rChild]):
                        nums[lChild], nums[parentIndex] = nums[parentIndex], nums[lChild]
                        index = lChild
                    else:
                        nums[rChild], nums[parentIndex] = nums[parentIndex], nums[rChild]
                        index = rChild
                else:
                    break


        def heapifyUp(index):

            while(index != 0):
                parentIndex = (index - 1) // 2

                if(nums[parentIndex] > nums[index]):
                    nums[parentIndex], nums[index] = nums[index], nums[parentIndex]

                index = parentIndex


        nums[ind] = val

        lChild = (2 * ind) + 1
        rChild = (2 * ind) + 2

        # finding if we need heapify up or heapify down
        if(ind == 0):
            # We are sure that we need to preform heapify down
            heapifyDown(ind)
        elif(lChild < n and nums[lChild] < nums[ind] or rChild < n and nums[rChild] < nums[ind]):
            heapifyDown(ind)
        else:
            parent = (ind - 1) // 2
            if(nums[ind] < nums[parent]):
                heapifyUp(ind)

        return nums

```

```python
# Build a min heap from a given Array
"""
Note - we can avoid all the leaf nodes in our for loop, check the lecture!!!!!
"""

class Solution:
    def checkAndReheapify(self, nums, index, n):
        lChild = (2 * index) + 1
        rChild = (2 * index) + 2


        if(index >= n or lChild >= n and rChild >= n):
            return

        swappedIndex = -1


        if(lChild < n and nums[lChild] < nums[index]):
            if(rChild < n and nums[rChild] < nums[lChild]):
                nums[rChild], nums[index] = nums[index], nums[rChild]
                swappedIndex = rChild
            else:
                nums[lChild], nums[index] = nums[index], nums[lChild]
                swappedIndex = lChild
        elif(rChild < n and nums[rChild] < nums[index]):
            if(lChild < n and nums[lChild] < nums[rChild]):
                nums[lChild], nums[index] = nums[index], nums[lChild]
                swappedIndex = lChild
            else:
                nums[rChild], nums[index] = nums[index], nums[rChild]
                swappedIndex = rChild

        if(swappedIndex != -1):
            return self.checkAndReheapify(nums, swappedIndex, n)
        else:
            return



    def buildMinHeap(self, nums):
        n = len(nums)
        index = n - 1

        if(n == 1):
            return nums

        for i in range(n-1, -1, -1):
            self.checkAndReheapify(nums, i, n)

        return nums
```

```python
#Max Heap

class Solution:

    def heapifyDown(self, index):
        n = len(self.heap)
        parent = index
        lChild = (index * 2) + 1
        rChild = (index * 2) + 2

        if(index >= n or lChild >= n and rChild >= n):
            return

        if(lChild > n and self.heap[lChild] > self.heap[parent]):
            if(rChild > n and self.heap[rChild] > self.heap[lChild]):
                self.heap[rChild], self.heap[parent] = self.heap[parent], self.heap[rChild]
                index = rChild
                return self.heapifyDown(index)
            else:
                self.heap[lChild], self.heap[parent] = self.heap[parent], self.heap[lChild]
                index = lChild
                return self.heapifyDown(index)

        elif(rChild > n and self.heap[rChild] > self.heap[parent]):
            if(lChild > n and self.heap[lChild] > self.heap[rChild]):
                self.heap[lChild], self.heap[parent] = self.heap[parent], self.heap[lChild]
                index = lChild
                return self.heapifyDown(index)
            else:
                self.heap[rChild], self.heap[parent] = self.heap[parent], self.heap[rChild]
                index = rChild
                return self.heapifyDown(index)
        else:
            return

    def heapifyUp(self, index):
        if(index == 0):
            return
        parent = (index - 1) // 2

        if(self.heap[index] > self.heap[parent]):
            self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
            return self.heapifyUp(parent)
        else:
            return



    def initializeHeap(self):
        self.heap = []


    def insert(self, key):
        self.heap.append(key)

        # we need to chck for heapify up condition
        root = self.getMax()
        curHeapSize = self.heapSize() - 1
        # print(f"Cur Heap size : {curHeapSize}")
        parentIndex = (curHeapSize - 1) // 2
        # print(f"Parent Index : {parentIndex}")

        if(key > root):
            # swap it with root and check the condition for heapify up at self.heapSize()
            self.heap[0], self.heap[curHeapSize] = self.heap[curHeapSize], self.heap[0]
            self.heapifyUp(curHeapSize)
        elif(key > self.heap[parentIndex]):
            self.heapifyUp(curHeapSize)

    def changeKey(self, index, new_val):
        # condition for heapify up
        if(new_val > self.heap[index]):
            self.heap[index] = new_val
            self.heapifyUp(index)
        elif(new_val < self.heap[index]):
            self.heap[index] = new_val
            self.heapifyDown(index)
        else:
            self.heap[index] = new_val


    def extractMax(self):
        curSize = self.heapSize()

        if(curSize == 1):
            self.heap.pop()

        elif(curSize != 0):

            # print(f"Initial Heap : {self.heap}")
            lastEle = self.heap.pop()
            # print(f"After popping : {self.heap}")
            self.heap[0] = lastEle
            self.heapifyDown(0)


    def isEmpty(self):
        if(self.heapSize() == 0):
            return 1
        return 0


    def getMax(self):
        return self.heap[0]

    def heapSize(self):
        return len(self.heap)
```

```python
# Check if an array represents a min heap

class Solution:
    def isHeap(self, nums):
        # we do this to make sure we run our loop for only parent node index, the belwo formula helps us to find that
        numsSiz = len(nums)
        n = (numsSiz // 2) - 1

        for i in range(n + 1):
            lChild = (i * 2) + 1
            rChild = (i * 2) + 2



            if(lChild < numsSiz):
                if(nums[lChild] < nums[i]):
                    # print(f"For parent index : {i}, checking lChild : {lChild} condition {nums[lChild]} < {nums[i]}")
                    return False
            if(rChild < numsSiz):
                if(nums[rChild] < nums[i]):
                    return False


        return True
```

```python
# Convert Min Heap to Max Heap

class Solution:
    def minToMaxHeap(self, nums):
        arrLen = len(nums)
        n = (arrLen // 2) - 1

        def heapifyDown(index):
            if(index > n):
                return
            print(f"Inside the sub function, for index : {index}")

            lChild = (index * 2) + 1
            rChild = (index * 2) + 2

            if(lChild < arrLen and nums[lChild] > nums[index]):
                if(rChild < arrLen and nums[rChild] > nums[lChild]):
                    # print(f"swapping with r Chld")
                    nums[rChild], nums[index] = nums[index], nums[rChild]
                    heapifyDown(rChild)
                else:
                    nums[lChild], nums[index] = nums[index], nums[lChild]
                    heapifyDown(lChild)

            elif(rChild < arrLen and nums[rChild] > nums[index]):
                nums[rChild], nums[index] = nums[index], nums[rChild]
                heapifyDown(rChild)
            else:
                return

        for i in range(n, -1, -1):
            heapifyDown(i)

        return nums
```

```python
# Heap sort
"""
TC to get max heap is O(n) (need to know how !!)
While loop O(n)
reheapify O(log n)
so total = O(n log n)
SC = O(1)
"""
class Solution:
    def heapSort(self, nums):
        arrLen = len(nums)

        till = (arrLen // 2) - 1

        def heapifyDown(index, n):


            lChild = (index * 2) + 1
            rChild = (index * 2) + 2

            largestInd = index

            # check if lChild have a valid index and is greater then the parent
            if(lChild <= n and nums[lChild] > nums[largestInd]):
                largestInd = lChild

            # check if rChild have a valid index and is greater then the parent
            if(rChild <= n and nums[rChild] > nums[largestInd]):
                largestInd = rChild

            # check if there is a greater value in either lChild or lChild if yes we swap
            if(largestInd != index):
                # we swap the parent node with the largest child
                nums[index], nums[largestInd] = nums[largestInd], nums[index]
                heapifyDown(largestInd, n)
            if(n == 2):
                print(f"parent : {index}, n : {n} lChild :{lChild}, rChild : {rChild}, swapped with {largestInd}")
            return


        # first we will convert the array to max heap
        def maxHeap(parIndexs):
            for i in range(parIndexs, -1, -1):
                heapifyDown(i, arrLen - 1)
            return

        # Step - 1 converting our array to max heap
        maxHeap(till)
        print(f"Max heap : {nums}")

        last = arrLen - 1

        # here we keep on extracting max element which is (nums[0] and swapping with the last node, and then we will ignore that node using last)

        while(last >= 0):
            nums[0], nums[last] = nums[last], nums[0]
            print(f"after Swap our array : {nums}")
            last -= 1
            heapifyDown(0, last)
            print(f"after heapify down our array : {nums}")

        return nums
```

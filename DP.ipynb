```python
# Simple recursion
class Solution:
    def fib(self, n: int) -> int:
        if(n <= 1):
            return n

        return self.fib(n - 1) + self.fib(n - 2)
```

```python
sol = Solution()

sol.fib(40)
```

```python
#solution with dp [Memoization]
class Solution:
    def fibDPMemoization(self, n):
        dpArr = [-1] * (n + 1)

        def recFib(n):
            if(n <= 1):
                return n
            if(dpArr[n] != -1):
                return dpArr[n]

            dpArr[n] = recFib(n - 1) + recFib(n - 2)
            return dpArr[n]

        return recFib(n)
```

```python
sol = Solution()

sol.fibDPMemoization(50)
```

```python
# solution with dp [tabulation]

class Solution:
    def fibDPTabulation(self, n):
        curFib = 1
        prev = 1
        prev2 = 0

        for i in range(2, n + 1):
            curFib = prev + prev2
            prev2 = prev
            prev = curFib

        return prev
```

```python
sol = Solution()

sol.fibDPTabulation(50)
```

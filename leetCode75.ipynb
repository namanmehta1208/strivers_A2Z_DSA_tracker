```python
#Merge Strings Alternately
def mergeAlternately(word1, word2):
  word1Len = len(word1)
  word2Len = len(word2)

  maxLen = max(word1Len, word2Len)
  ans = ""

  for i in range(maxLen):
    if(i < word1Len):
      ans += word1[i]

    if(i < word2Len):
      ans += word2[i]

  return ans


```

```python
#1071. Greatest Common Divisor of Strings
"""
So here we tried a simple approach where we took the lengths of the string 1
and 2 then we used a greedy approach where we started with the complete string2
did a mod with both the strings confirming that the selected string length will
divide both of them and then for the selected length we got the remainder when
we will divide it with the first string and also the string2 and then using the
remainder we will multiply the selcted GCD for both string1 and string2
if the condition satisfies for both we found our GCD else we will go to the
next iteration.
"""

def gcdOfStrings(str1, str2):
  str1Len = len(str1)
  str2Len = len(str2)
  i = str2Len
  gcdChar = ""

  while(i > 0):
    gcdChar = str2[0:i]
    print(gcdChar)

    str1Mod = str1Len % len(gcdChar)
    str2Mod = str2Len % len(gcdChar)

    if(str1Mod == 0 and str2Mod == 0):
      rem1 = str1Len // len(gcdChar)
      rem2 = str2Len // len(gcdChar)
      if(gcdChar * rem1 == str1 and gcdChar * rem2 == str2):
        print(gcdChar * rem1 == str1 and gcdChar * rem2 == str2)
        return gcdChar

    i -= 1

  return ""




```

```python
#1431. Kids With the Greatest Number of Candies
def kidsWithCandies(candies, extraCandies):
  maxCandies = 0
  result = []

  for i in candies:
    maxCandies = max(maxCandies, i)

  for j in range(0, len(candies)):
    if((candies[j] + extraCandies) >= maxCandies):
      result.append(True)
    else:
      result.append(False)

  return result
```

```python
#151. Reverse Words in a String
def reverseWords(s):
  wordList = []
  eachWord = ""
  resultStr = ""

  for n in range(0, len(s)):
    if(s[n] != " "):
      # print(s[n])
      eachWord += s[n]
      # print(n)
    elif(eachWord != ""):
      wordList.append(eachWord)
      eachWord = ""

  if(eachWord != ""):
    wordList.append(eachWord)

  for word in range(len(wordList) -1, -1, -1):
    if(word > 0):
      resultStr += (wordList[word] + " ")
    else:
      resultStr += wordList[word]


  return resultStr
```

```python
# 605. Can Place Flowers
def canPlaceFlowers(flowerbed, n):
  i = 0
  k = len(flowerbed)

  while(i < k and n != 0):
    if(i == 0):
      if(flowerbed[i] == 0 and flowerbed[i + 1] == 0):
        flowerbed[i] = 1
        print('plotted at', i)
        n -= 1
        i += 1
      else:
        i += 1

    elif(i == (k - 1)):
      if(flowerbed[i] == 0 and flowerbed[i - 1] == 0):
        flowerbed[i] == 1
        print('plotted at', i)
        n -= 1
        break
      else:
        break

    else:
      if(flowerbed[i - 1] == 0 and flowerbed[i] == 0 and flowerbed[i + 1] == 0):
        flowerbed[i] = 1
        print('plotted at', i)
        n -= 1
        i += 1
      else:
        i += 1

  if(n == 0):
    return True
  else:
    return False

```

```python
# 238. Product of Array Except Self
def productExceptSelf(nums):
  n = len(nums)
  output = []
  pre = 1
  post = 1

  for i in range(0, n):
    output.append(pre)
    pre *= nums[i]

  # print('Output after first pass = ', output)

  for j in range(n - 1, -1, -1):
    output[j] = output[j] * post

    post *= nums[j]

  return output
```

```python
# 334. Increasing Triplet Subsequence
"""
so this problem can be salved in one pass using the following
approach, so we maintain two variables num1 and num2. num1 will
have the lowest number and num2 will have the second lowest number
if a number is greater than both num1 and num2 then we have found
our truplet
"""
def increasingTriplet(nums):
  num1 = float('inf')
  num2 = float('inf')

  for n in nums:
    if n <= num1:
      num1 = n
    elif n <= num2:
      num2 = n
    else:
      return True

  return False
```

```python
# 283. Move Zeroes
def moveZeroes(nums):
  """
  ok so we are solving it with one pass, we keep the
  track of 0 using j and as soom as we find a non zero
  element we swap and then increment the j as well
  """
  n = len(nums)
  i = 0
  j = 0

  if(n == 1):
    return nums

  for i in range(0, n):
    if(nums[i] != 0):
      nums[i], nums[j] = nums[j], nums[i]
      j += 1

  return nums
```

```python
# 392. Is Subsequence
def isSubsequence(s, t):
  lens = len(s)
  lent = len(t)
  totMatch = 0
  i, j = 0, 0

  while(i < lens and j < lent):
    if(s[i] == t[j]):
      totMatch += 1
      i += 1
      j += 1
    else:
      j += 1

  if(totMatch == lens):
    return True
  else:
    return False
```

```python
# 1679. Max Number of K-Sum Pairs
def maxOperations(nums, k):
  n = len(nums)
  i = 0
  j = n - 1
  count = 0

  nums.sort()

  while(i < j):
    sum = nums[i] + nums[j]
    if(sum == k):
      count += 1
      i += 1
      j -= 1
    elif(sum > k):
      j -= 1
    else:
      i += 1

  return count
```

```python
# 643. Maximum Average Subarray I
"""
we create two pointer i, j starting from 0 then we create a window whose
size is equal to k and we start calculating the average moving the window
with each calculation and keeping the maximum average we saw throught the
traversal.
"""
def findMaxAverage(nums, k):
  i, j = 0, 0
  sum = 0
  maxSum = float('-inf')

  if(len(nums) == 1):
    return sum

  while(j < len(nums)):
    sum += nums[j]

    if((j - i) + 1 == k):
      avg = sum / k
      print('average of i = ', i, ' j = ', j, ' = ', avg)
      maxSum = max(avg, maxSum)
      j += 1
      sum -= nums[i]
      i += 1
    else:
      j += 1

  return maxSum

```

```python
# 1456. Maximum Number of Vowels in a Substring of Given Length
def maxVowels(s, k):
  i, j = 0, 0
  set = {'a', 'e', 'i', 'o', 'u'}
  curCount = 0
  maxCount = 0

  if(len(s) == 1):
    if(s[0] in set):
      return 1
    else:
      return 0

  while(j < len(s)):
    if(s[j] in set):
      curCount += 1
    if((j - i) + 1 == k):
      maxCount = max(curCount, maxCount)
      if(s[i] in set):
        curCount -= 1
      i += 1
      j += 1
    else:
      j += 1

  return maxCount
```

```python
"""
Better Version
so in this version as we are exeeding number of zeros we are
reducing the window length from the left till the number of
zeros becaames less than k so overall we are not going back
in out traversal where as in the above solution we were going back
"""
def longestOnesV2(nums, k):
  zerosCounter = 0
  maxOnes = 0
  l = 0

  for i in range(len(nums)):
    if(nums[i] == 0):
      zerosCounter += 1

    while(zerosCounter > k):
      if(nums[l] == 0):
        zerosCounter -= 1
      l += 1

    maxOnes = max((i - l) + 1, maxOnes)

  return maxOnes
```

```python
# 1493. Longest Subarray of 1's After Deleting One Element
"""
so we expand our window and as soon as we get two zweros in our
window we shrink from i until we get rid of one zero and so onn
:D
"""
def longestSubarray(nums):
  i, j = 0, 0
  maxOnes = 0
  zeroCounter = 0

  for j in range(len(nums)):
    if(nums[j] == 0):
      zeroCounter += 1
    if(zeroCounter == 2):
      while(nums[i] != 0):
        i += 1
      zeroCounter -= 1
      i += 1

    maxOnes = max((j - i), maxOnes)


  return maxOnes
```

```python
# 724. Find Pivot Index
def pivotIndex(nums):
  total = 0
  rightSum = 0
  leftSum = 0

  for i in range(0, len(nums)):
    total += nums[i]

  rightSum = total

  for i in range(0, len(nums)):
    rightSum -= nums[i]
    # print('at Index ', i,  ' right sum = ', rightSum, ' left sum = ', leftSum)

    if(rightSum == leftSum):
      return i

    leftSum += nums[i]

  return -1
```

```python
# 2215. Find the Difference of Two Arrays
def findDifference(num1, num2):
  numSet = set()
  num1Diff = []
  num2Diff = []

  for num in num2:
    numSet.add(num)

  for num in num1:
    if(num not in numSet and num not in num1Diff):
      num1Diff.append(num)

  numSet.clear()

  for num in num1:
    numSet.add(num)

  for num in num2:
    if(num not in numSet and num not in num2Diff):
      num2Diff.append(num)

  return [num1Diff, num2Diff]
```

```python
# 1207. Unique Number of Occurrences
"""
so firstly we are using the hash table to store
the count for each nnumber then using hash.values()
we are storing these values in a set which would allow
us to avoid duplicates now if the length of the set
and hash.values differ it means that there are numbers
with the same count!!!!
"""
def uniqueOccurrences(arr):
  counter = {}

  for num in arr:
    counter[num] = counter.get(num, 0) + 1

  values = counter.values()
  uniqCount = set(values)

  return len(counter) == len(uniqCount)
```

```python
# 2352. Equal Row and Column Pairs
from collections import defaultdict

def equalPairs(grid):
  rowsCounts = defaultdict(int)
  counter = 0

  for rows in grid:
    rowsCounts[tuple(rows)] += 1

  for columns in zip(*grid):
    counter += rowsCounts[columns]

  return counter
```

```python
# 394. Decode String
def decodeString(s):

  stack = []

  for i in range(len(s)):
    if(s[i] != "]"):
      stack.append(s[i])
    else:
      substr = ""
      while(stack[-1] != "["):
        substr = stack.pop() + substr
      stack.pop()

      k = ""
      while(stack and stack[-1].isdigit()):
        k = stack.pop() + k

      stack.append(int(k) * substr)

  return "".join(stack)
```

```python
# 933. Number of Recent Calls
"""
Simple initialize the queue then once we get a new time we append it
and also calculate the range using t - 3000 if anything is less
than this range remove it from the front and adjust our count based
on that!!!
"""
from collections import deque

class RecentCounter:

    def __init__(self):
      self.q = deque()
      self.counter = 0


    def ping(self, t: int) -> int:
      self.q.append(t)
      self.counter += 1

      while self.q and self.q[0] < (t - 3000):
        self.q.popleft()
        self.counter -= 1

      return self.counter
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def oddEvenList(self, head):
      oddLL = head
      evenLL = head.next
      evenHead = evenLL

      while(evenLL and evenLL.next):
        # print("oddLL pointing ", oddLL.val, " to ", evenLL.next.val)
        oddLL.next = evenLL.next
        oddLL = oddLL.next
        # print("oddLL has now become ", oddLL.val)

        # print("evenLL pointing ", evenLL.val, " to ", oddLL.next.val)
        evenLL.next = oddLL.next
        evenLL = evenLL.next
        # print("evenLL has now become ", evenLL.val)

      oddLL.next = evenHead

      return head
```

```python
# 206. Reverse Linked List
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head):

      if(head and head.next):
        initialNode = head
        secondNode = head.next
        initialNode.next = None
        while(secondNode and secondNode.next):
          ref = secondNode.next
          secondNode.next = initialNode
          initialNode = secondNode
          secondNode = ref
        head = secondNode
        head.next = initialNode
        return head
      else:
        return
```

```python
# 104. Maximum Depth of Binary Tree
# Definition for a binary tree node.
from collections import deque
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:

    # using recursion
    # def maxDepth(self, root):
    #   if not root:
    #     return 0

    #   return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))

    # using BFS (queue)
    def maxDepth(self, root):
      if not root:
        return 0

      count = 0
      q = deque([root])

      while q:
        for i in range(len(q)):
          node = q.popleft()

          if(node.left):
            q.append(node.left)

          if(node.right):
            q.append(node.right)
        count += 1
      return count
```

```python
"""
So initialy I used queue here to traverse the tree in BFS mode
but we faced a problem as the right side of the tree was smaller
in root1 we got it's node with no child first which made out order
incorrect so then I went with stack which used DFS which forced us to
process the left hand side first ensuring our order stays correct!!
"""
from collections import deque
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def leafSimilar(self, root1, root2):
      # if(not root1 or not root2):
      #   return 0

      s1 = [root1]
      s2 = [root2]

      list1 = []
      list2 = []

      while s1:
        node = s1.pop()

        if(node.right):
          s1.append(node.right)
        if(node.left):
          s1.append(node.left)

        if(not node.left and not node.right):
          list1.append(node.val)

      while s2:
        node = s2.pop()

        if(node.right):
          s2.append(node.right)
        if(node.left):
          s2.append(node.left)

        if(not node.left and not node.right):
          list2.append(node.val)

      # return (list1 == list2)
      return list1, list2
```

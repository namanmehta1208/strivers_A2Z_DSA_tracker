```python
#Traversal Techniques
from collections import deque
class Solution:
    def dfsOfGraph(self, V, adj):
        trvs = []
        visited = set()
        def dfs(curV, curAdj):
            if(curV in visited):
                return
            trvs.append(curV)
            visited.append(curV)

            for i in curAdj:
                dfs(i, adj[i])

        dfs(0, adj[0])
        print(f"trvs : {trvs}")
        return trvs


    def bfsOfGraph(self, V, adj):
        visited = set()
        trvs = []
        que = deque([0])

        visited.add(0)

        while(que):
            curNode = que.popleft()
            trvs.append(curNode)

            linkNodes = adj[curNode]
            for i in linkNodes:
                if(i not in visited):
                    que.append(i)
                    visited.add(i)
            print(f"Current Que after append : {que}")
        return trvs
```

```python
"""
here for each node we are using BFS to reaverse it and putting it in visited array
one important thing we are doing here is using edges to create adjacency list
using it we are performing BFS and counting vertices connected(forming a graph) or alone
"""

from collections import deque
# Connected Components
class Solution:
    def findNumberOfComponent(self, V, edges):
        visited = [0] * V
        count = 0
        adjLi = [[] for _ in range(V)]

        def bfsTraversal(vertex):
            que = deque([vertex])

            while(que):
                curVertex = que.popleft()
                linkedNodes = adjLi[curVertex]

                for i in linkedNodes:
                    if(visited[i] != 1):
                        que.append(i)
                        visited[i] = 1



        # Creating adjacency list
        for i in range(len(edges)):
            adjLi[edges[i][0]].append(edges[i][1])
            adjLi[edges[i][1]].append(edges[i][0])

        print(f"adj : {adjLi}")
        for i in range(V):
            if(visited[i] != 1):
                bfsTraversal(i)
                count += 1

        return count
```

```python
from collections import deque
# Number of provinces
class Solution:
    def numProvinces(self, adj):
        visited = [0] * len(adj)
        count = 0

        def bfsTraversal(vertex):
            que = deque()
            que.append(vertex)
            visited[vertex] = 1

            while(que):
                curVertex = que.popleft()
                # print(f"CurVertex : {curVertex}")
                curAdj = adj[curVertex]

                for i in range(len(curAdj)):
                    if(curAdj[i] == 1 and visited[i] != 1):
                        que.append(i)
                        visited[i] = 1


        for i in range(len(adj)):
            if(visited[i] != 1):
                bfsTraversal(i)
                count += 1
        return count
```

```python
# Flood fill algorithm
from collections import deque
class Solution:
    def floodFill(self, image, sr, sc, newColor):
        rowSiz = len(image)
        colSiz = len(image[0])

        print(f"len(image : {len(image)}, len (image[0]) : {len(image[0])})")
        rowNeigh = [0, -1, 0, 1]
        colNeigh = [1, 0, -1, 0]
        sourceColor = image[sr][sc]

        def validatePos(row, col):
            if(row < 0 or row >= rowSiz):
                return False
            if(col < 0 or col >= colSiz):
                return False

            return True

        def bfsTraversal(curRow, curCol):
            image[curRow][curCol] = newColor

            for i in range(4):
                newRow = curRow + rowNeigh[i]
                newCol = curCol + colNeigh[i]

                # print(f"{image[newRow][newCol]}, {sourceColor}")

                if(validatePos(newRow, newCol) and image[newRow][newCol] == sourceColor and image[newRow][newCol] != newColor):
                    print("Condition is True")
                    image[newRow][newCol] = newColor
                    bfsTraversal(newRow, newCol)

        bfsTraversal(sr, sc)

        return image
```

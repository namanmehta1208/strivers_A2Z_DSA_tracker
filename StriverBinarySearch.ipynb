```python
# Find square root of a number
class Solution:
    def floorSqrt(self, n: int) -> int:
        i, j = 1, n

        while(i <= j):
            mid = (i + j) // 2
            sqr = mid * mid

            if(sqr <= n):
                ans = mid
                i = mid + 1
            elif(sqr > n):
                j = mid - 1


        return ans
```

```python
sol = Solution()

sol.floorSqrt(36)
```

```python
# Find Nth root of a number
class Solution:
    def NthRoot(self, n, m):
        ans = -1

        i, j = 1, m

        while(i <= j):
            mid = (i + j) // 2
            val = mid ** n

            if(val == m):
                ans = mid
                break
            elif(val > m):
                j = mid - 1
            elif(val < m):
                i = mid + 1

        return ans
```

```python
sol = Solution()
sol.NthRoot(3, 27)
```

```python
# Find the smallest divisor
import math
class Solution:
    def smallestDivisor(self, nums, limit):

        def division(num):
            div = 0

            for i in nums:
                div += math.ceil(i / num)
            print(f"for {num} the div is : {div}")
            return div




        ans = 0

        i, j = 1, max(nums)

        while(i <= j):
            mid = (i + j) // 2

            if(division(mid) <= limit):
                ans = mid
                j = mid - 1
            else:
                i = mid + 1

        return ans
```

```python
sol = Solution()

sol.smallestDivisor([8,4,2,3], 4)
```

```python
# Koko eating bananas
class Solution:
    def minimumRateToEatBananas(self, nums, h):
        res = 0
        def totHour(rate):
            ans = 0
            for i in nums:
                ans += math.ceil(i / rate)
            return ans

        i, j = 1, max(nums)

        while(i <= j):
            mid = (i + j) // 2

            if(totHour(mid) <= h):
                res = mid
                j = mid - 1
            else:
                i = mid + 1
        return res
```

```python
sol = Solution()
sol.minimumRateToEatBananas([25, 12, 8, 14, 19], 5)
```
